```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data= np.random.randint(1, 50,100)
```

## Histograma

```python
plt.hist(data)
plt.show()
```

```python
plt.hist(data,bins=50,histtype='bar')
```

```python
plt.hist(data,bins=50,histtype='step')
plt.show()
```

## Boxplot

```python
plt.boxplot(data)
plt.show()
```

```python
data= np.append(data,200)
plt.boxplot(data,vert=False,patch_artist=True,notch=True)
plt.show()
```

```python
data= np.append(data,200)
plt.boxplot(data,vert=False,patch_artist=True,notch=True,showfliers=False)
plt.show()
```

### Scatter

```python
N=50
X= np.random.rand(N)
Y= np.random.rand(N)
area = (30 * np.random.rand(N))**2
colors = np.random.rand(N)
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
plt.scatter(X,Y,s=area,c=colors)
plt.show()
```

```python
plt.scatter(X,Y,s=area,c=colors,marker='x')
plt.show()
```

```python
plt.scatter(X,Y,s=area,c=colors,marker='p')
plt.show()
```

```python
plt.scatter(X,Y,s=area,c=colors,marker='<',alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=np.sin(x)
```

```python
fig , axes = plt.subplots()
axes.plot(x,y,'b')
```

```python
fig , axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,'b:')
axes[1].plot(y,x,'r--')
```

```python
fig , (ax1,ax2) = plt.subplots(nrows=1,ncols=2)
ax1.plot(x,y,'b:')
ax2.plot(y,x,'r--')
```

```python
fig , axes = plt.subplots(nrows=2,ncols=4)
axes[0,0].plot(x,y,'b:')
axes[0,1].plot(x,np.cos(x),'y:')
axes[0,2].plot(x,np.tan(x),'r-')
axes[0,3].plot(x,np.sin(x),'g:')
axes[1,0].plot(y,x,'r--')

```

```python
fig , ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(nrows=2,ncols=4)
ax1.plot(x,y,'b:')
ax2.plot(x,np.cos(x),'y:')
ax3.plot(x,np.tan(x),'r-')
ax4.plot(x,np.sin(x),'g:')
ax5.plot(y,x,'r--')
fig.tight_layout()
```

```python
fig , axes = plt.subplots(nrows=2,ncols=4)
axes[0,0].plot(x,y,'b:')
axes[0,1].plot(x,np.cos(x),'y:')
axes[0,2].plot(x,np.tan(x),'r-')
axes[0,3].plot(x,np.sin(x),'g:')

axes[1, 0].plot(y, x, 'r--')
axes[1, 1].pie([1, 2, 3], labels=['A', 'B', 'C'], autopct='%1.1f%%', colors=['red', 'green', 'blue'])
axes[1, 2].hist(x, bins='auto', color='b')
axes[1, 3].plot(y, x, 'r--')

```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=np.sin(x)
```

```python
print(plt.style.available)
```

```python
plt.style.use('dark_background')
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,'r--')
ax.plot(x,x+2,'b-')
ax.plot(x,x+3,'g.-')
ax.plot(x,x+4,'y:')
```

##Pyplot

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,color='green',alpha=0.5,linewidth=5,marker='o',markersize=8,markerfacecolor="#CCFFCC")
ax.plot(x,x+2,color='blue',linewidth=8,linestyle='-',marker='x')
ax.plot(x,x+3,color='#66FF66',linestyle=':',marker='v')
ax.plot(x,x+4,color='#CCFFCC',linestyle='--',marker='P')
```

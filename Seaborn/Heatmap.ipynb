```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
tips =sns.load_dataset('tips')
tips.head(2)
```

```python
tips.corr(numeric_only=True)  
```

```python
sns.heatmap(tips.corr(numeric_only=True)  )
plt.show()
```

```python
sns.heatmap(tips.corr(numeric_only=True),annot=True  )
plt.show()
```

```python
sns.heatmap(tips.corr(numeric_only=True),annot=True,cmap='coolwarm',linewidths=5,linecolor= 'black',vmin=0.5,vmax=1,cbar=False )
plt.show()
```

```python
sns.heatmap(tips.corr(numeric_only=True),annot=True,cmap='coolwarm',linewidths=5,linecolor= 'black',vmin=0.5,vmax=1,cbar=True )
plt.show()
```
